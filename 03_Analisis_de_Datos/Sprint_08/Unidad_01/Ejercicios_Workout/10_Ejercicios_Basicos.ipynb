```python
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np
```

### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
x = np.linspace(0,10,100)
fig = plt.figure()
plt.subplot(1,2,1,)
plt.plot(x, np.sin(x), label='sin(x)')
plt.ylim(-1.1, 1.1)
plt.xlim(0, 10.5)
plt.xticks(np.arange(0,12,2.5))
plt.yticks(np.arange(-1,1.1,0.25))
plt.subplot(1,2,2)
plt.plot(x, np.cos(x), label='cos(x)')
plt.ylim(-1.1,1.1)
plt.xlim(0, 10.5)
plt.xticks(np.arange(0,12,2.5))
plt.yticks(np.arange(-1,1.1,0.25));
```

### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

```python
#Con axes
fig,ax = plt.subplots(3)
ax[0].plot(x, np.sin(x), label='sin(x)',color="red",lw=2)  
ax[0].set_ylim(-1.1, 1.1)  
ax[0].set_yticks(np.arange(-1,1.1,1))
ax[1].plot(x, np.cos(x), label='cos(x)',color="g",lw=2)
ax[1].set_ylim(-1.1, 1.1)
ax[1].set_yticks(np.arange(-1,1.1,1))
ax[2].plot(x, np.tan(x), label='tan(x)',color="blue",lw=2)
ax[2].set_yticks(np.arange(-40,21,20));
```

```python
fig = plt.figure(figsize=(8, 6))
plt.subplot(3,1,1)
plt.plot(x, np.sin(x), label='sin(x)',color="r",lw=2)
plt.yticks(np.arange(-1,1.1,1))
plt.ylim(-1.1,1.1)

plt.subplot(3,1,2)
plt.plot(x, np.cos(x), label='cos(x)',color="g",lw=2)
plt.yticks(np.arange(-1,1.1,1))
plt.ylim(-1.1,1.1)

plt.subplot(3,1,3)
plt.plot(x, np.tan(x), label='tan(x)',color="b",lw=2)
plt.yticks(np.arange(-40,21,20));
```

### Ejercicio 4

Reproduce la figura siguiente. Recuerda que las funciones seno hiperblico y coseno hiperblico se obtienen as:
* np.sinh, seno hiperblico;
* np.cosh, coseno hiperblico;


```python
fig,ax = plt.subplots(2,1)
fig.subplots_adjust(hspace=0.5)
ax[0].plot(x, np.sinh(x),color="r",lw=2)
ax[0].set_title("Seno Hiperblico")
ax[0].set_ylabel("sinh(x)")
ax[0].set_xlabel("x")
ax[0].set_ylim(-600,12000)
ax[0].set_xlim(-.5,10.5)

ax[1].plot(x, np.cosh(x),label="Coseno Hiperblico",color="b",lw=2)
ax[1].set_title("Coseno Hiperblico")
ax[1].set_ylabel("cosh(x)")
ax[1].set_xlabel("x")
ax[1].set_xlim(-.5,10.5)
ax[1].set_ylim(-600,12000)
```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosicin (tal y como ves en la figura), emplea `subplots_adjust` y los parmetros hspace y wspace.

```python
fig,ax = plt.subplots(2,2)
fig.subplots_adjust(hspace=0.5,wspace=0.5)
ax[0,0].plot(x,np.sin(x))
ax[0,0].set_ylim(-1.1,1.1)

ax[0,1].plot(x,np.cos(x))
ax[0,1].set_ylim(-1.1,1.1)

ax[1,0].plot(x,np.tan(x))
ax[1,0].set_yticks(np.arange(-40,21,20))

ax[1,1].plot(x,np.log(x));
```
