### Ejercicio 5

Realiza una anlisis de todas las posibles correlaciones dos a dos, de forma numrica primero y luego empleado un heatmap y un pairplot de seaborn. Qu variables tienen o aparentemente tienen una mayor correlacin con "total?

```python
corr_matrix = df_crashes.corr(numeric_only=True)
corr_matrix
```

```python
plt.figure(figsize=(10, 8))  # Ya lo veremos pero esto permite ajustar el tamao de las grficas
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", 
            cbar=True, square=True, linewidths=.5) # el cmap es el rango de colores usado para representar "el calor"

plt.title('Matriz de Correlacin')
plt.xticks(rotation=60)  # Rota las etiquetas de las x si es necesario
plt.yticks(rotation=45)  # Rota las etiquetas de las y si es necesario

plt.show()
```
